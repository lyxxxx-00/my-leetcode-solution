```python
#!/bin/python3

import math
import os
import random
import re
import sys



#
# Complete the 'flippingMatrix' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY matrix as parameter.
#

def flippingMatrix(matrix):
    # Write your code here
    n = len(matrix) // 2
    sum = 0
    
    for i in range(n):
        for j in range(n):
            topLeft = matrix[i][j]
            topRight = matrix[i][2*n - j -1]
            bottomLeft = matrix[2*n - i -1][j]
            bottomRight = matrix[2*n - i -1][2*n - j -1]
            
            max_value = max(topLeft, topRight, bottomLeft, bottomRight)
            sum += max_value
    
    return sum

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        n = int(input().strip())

        matrix = []

        for _ in range(2 * n):
            matrix.append(list(map(int, input().rstrip().split())))

        result = flippingMatrix(matrix)

        fptr.write(str(result) + '\n')

    fptr.close()
```

### The variable names 

top-left, top-right, bottom-left, and bottom-right are not valid in Python due to the hyphens. Variable names should use underscores or no special characters.

### Mirrored Positions:

For each cell (i, j) in the top-left n x n quadrant, the possible mirrored positions in the matrix are:
Top-left: (i, j)
Top-right: (i, 2n-j-1)
Bottom-left: (2n-i-1, j)
Bottom-right: (2n-i-1, 2n-j-1)
These positions are considered because flipping the entire matrix (or specific quadrants) can transform each position in the top-left quadrant into one of these mirrored positions.

### Why Mirrored Positions are Sufficient:

By comparing the cell values in the mirrored positions, you are effectively covering all potential scenarios that could occur after any flipping operation. Each cell in the top-left quadrant can be transformed into its mirrored positions, and thus selecting the maximum from these positions ensures that you get the best possible value for each cell.
